```python
from bertviz import attention, visualization
import torch
from mptb import BertModel, Config
from mptb.tokenization import FullTokenizer
from IPython.display import display

tokenizer = FullTokenizer('data/multi_cased_L-12_H-768_A-12/vocab.txt')
model_cfg = Config.from_json('data/multi_cased_L-12_H-768_A-12/bert_config.json', len(tokenizer))
model = BertModel(model_cfg)
model.load_state_dict(torch.load('data/multi_cased_L-12_H-768_A-12/multi_cased_L-12_H-768_A-12.pt', map_location='cpu'))
```

```python
sentence_a = ""
sentence_b = ""
attention_visualizer = visualization.AttentionVisualizer(model, tokenizer)
tokens_a, tokens_b, attn = attention_visualizer.get_viz_data(sentence_a, sentence_b)
call_html()
attention.show(tokens_a, tokens_b, attn)
```
